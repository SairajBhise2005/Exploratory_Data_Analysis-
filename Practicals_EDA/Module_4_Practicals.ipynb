```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

data=fetch_california_housing()
print(data)
x=pd.DataFrame(data.data, columns=data.feature_names)
y=pd.Series(data.target, name='MedHouseVal')

x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2)

def forward_feature_selection(x_train, y_train, x_test, y_test):
    # Initialize features with the column names of x_train
    features = list(x_train.columns)
    selected_features=[]
    best_score=float('inf')

    # Rename the variable to avoid shadowing the function name
    current_score = best_score

    while features:
        scores=[]
        for feature in features:
          subset= selected_features+[feature]
          model=LinearRegression()
          model.fit(x_train[subset],y_train)
          y_pred=model.predict(x_test[subset])
          score=mean_squared_error(y_test,y_pred)
          scores.append((score,feature))
        scores.sort()
        best_score, best_feature=scores[0]

        # Compare current score to best score
        if best_score < current_score:
            selected_features.append(best_feature)
            features.remove(best_feature)
            current_score = best_score  # Update current best score
        else:
            break
    return selected_features

selected_features=forward_feature_selection(x_train, y_train, x_test, y_test)
print("Selected Features:", selected_features)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

iris = datasets.load_iris()
X = iris.data
y = iris.target

print("Shape of X (features):", X.shape)
print("Shape of y (target labels):", y.shape)

print("\nFirst 5 rows of X:\n", X[:5])
print("\nFirst 5 values of y:", y[:5])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Original means:", np.mean(X, axis=0))
print("Original standard deviations:", np.std(X, axis=0))
print("\nMeans after scaling:", np.mean(X_scaled, axis=0))
print("Standard deviations after scaling:", np.std(X_scaled, axis=0))

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
pca_df['Target'] = y
print(pca_df)

plt.figure(figsize=(8, 6))
colors = ['r', 'g', 'b']
targets = [0, 1, 2]

for target, color in zip(targets, colors):
    indices_to_keep = pca_df['Target'] == target
    plt.scatter(pca_df.loc[indices_to_keep, 'Principal Component 1'],
                pca_df.loc[indices_to_keep, 'Principal Component 2'],
                c=color,
                s=50,
                label=iris.target_names[target])

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA of IRIS Dataset")
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

df['target'] = iris.target
print('Dataset HEAD')
print(df.head())
print("\nSummary Statistics:")
print(df.describe())

# Function to bin data
def bin_data(column, bins, labels):
    return pd.cut(column, bins=bins, labels=labels, include_lowest=True)

# Bin numeric columns
bin_labels = ['Low', 'Medium', 'High']
for col in iris.feature_names:
    df[col + '_binned'] = bin_data(df[col], bins=3, labels=bin_labels)

print(df.head())

# Drop original numeric columns
df = df.drop(columns=iris.feature_names)
print(df.head())

# Encode binned columns
df_encoded = pd.get_dummies(df, columns=[col + '_binned' for col in iris.feature_names])
print("\nEncoded DataFrame:")
print(df_encoded.head())

# Validate binary encoding
assert df_encoded.applymap(lambda x: x in [0, 1]).all().all(), "Encoded DataFrame contains values other than 0 or 1."

# Perform Apriori analysis
min_support = 0.1
frequent_itemsets = apriori(df_encoded, min_support=min_support, use_colnames=True)
print("\nFrequent Itemsets:")
print(frequent_itemsets)

# Generate association rules
min_confidence = 0.6
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=min_confidence)

if not rules.empty:
    print("\nAssociation Rules:")
    print(rules)

    # Plot Support vs. Confidence
    plt.figure(figsize=(10, 6))
    plt.scatter(rules['support'], rules['confidence'], alpha=0.7, c="blue")
    plt.xlabel('Support')
    plt.ylabel('Confidence')
    plt.title('Support vs. Confidence')
    plt.grid(True)
    plt.show()
else:
    print("\nNo association rules were generated.")
```
