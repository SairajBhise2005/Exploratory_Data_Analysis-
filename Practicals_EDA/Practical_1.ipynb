```python
np.random.seed(42)
data=pd.DataFrame({
"Normal":np.random.normal(loc=0, scale=1, size=1000),
"Skewed": np.random.exponential(scale=2, size=1000),
"Bimodal": np.concatenate([np.random.normal(loc=0,scale=0.5, size=500),np.random.normal(loc=0, scale=1, size=500)]),
"Uniform": np.random.uniform(low=0, high=10, size=1000)
})
```

```python
fig, axes=plt.subplots(2,2, figsize=(14,10))

axes[0,0].hist(data['Normal'], bins=30, color='blue', alpha=0.7)
axes[0,0].set_title("Normal Distribution")

axes[0,1].hist(data['Skewed'], bins=30, color='green', alpha=0.7)
axes[0,1].set_title("Skewed Distribution")

axes[1,0].hist(data['Bimodal'], bins=30, color='red', alpha=0.7)
axes[1,0].set_title("Bimodal Distribution")

axes[1,1].hist(data['Uniform'], bins=30, color='purple', alpha=0.7)
axes[1,1].set_title("Uniform Distribution")

plt.tight_layout()

plt.show()
```

```python
import pandas as pd

data={
    'Date':['2030-01-01','2030-01-01','2030-02-02', '2030-02-02'],
    'Category':['A','B','A','B'],
    'Value':[10,20,30,40]
}

df=pd.DataFrame(data)
print('Original DataFrame:')
print(df)
```

```python
#filtering rows based on a condition
filtered_row=df[df['Value']>20]
print('\n Filtered Rows:')
print(filtered_row)
```

```python
#filtering using like paramter
selected_columns=df.filter(like='Date')
print("\n selected Columns:")
print(selected_columns)
```

```python
#Using regex Parameter
selected_columns_regex=df.filter(regex='^C|^V')
print("\n selected columns with regex:")
print(selected_columns_regex)
```

```python
#using columns based on a list
selected_columns_list=df.filter(items=['Date','Value'])
print("\n selected columns with list:")
print(selected_columns_list)
```

```python
import seaborn as sns
df=sns.load_dataset('titanic')
print(df.head())
print(df.dtypes)
```

```python
numerical_cols=df.select_dtypes(include=['int64','float64']).columns
print('Numerical_Columns:',numerical_cols.tolist())

categorical_cols=df.select_dtypes(include=['object','category','bool']).columns
print("Categorical Columns:", categorical_cols.tolist())
```

```python
def check_discrete_or_continuous(df,var):
    if all(df[var].apply(lambda x: float(x).is_integer())):
        print(f"Column '{var}' is discrete.")

    else:
        print(f"Column '{var}' is continuous.")

check_discrete_or_continuous(df,'survived')
check_discrete_or_continuous(df,'fare')
```

```python
def classify_nominal_ordinal(df,categorical_cols):
    ordinal_keywords=['level', 'rank', 'grade', 'stage','position', 'order','age','class']
    nominal_cols=[]
    ordinal_cols=[]

    for col in categorical_cols:
        if any(keywords in col.lower() for keywords in ordinal_keywords):
            ordinal_cols.append(col)
        else:
            nominal_cols.append(col)
    return nominal_cols, ordinal_cols

nominal_cols, ordinal_cols= classify_nominal_ordinal(df,categorical_cols)
print("Nominal Columns:", nominal_cols)
print("Ordinal Columns:", ordinal_cols)
```

```python
import matplotlib.pyplot as plt
import numpy as np

label=['Car', 'Bike', 'Truck', 'Cycle', 'Jeeps', 'Ambulance']
vehicle_no=[941, 854, 4595, 2125, 942, 509]

index= np.arange(len(label))
print('Total Labels:', len(label))
print("Indexing", vehicle_no)
plt.bar(index, vehicle_no)
plt.xlabel('Type', fontsize=15)
plt.ylabel('No of Vehicles', fontsize=15)
plt.xticks(index, label, fontsize=10)
plt.title('Market Share for each Genre 1995-2017')

plt.show()
plt.figure(figsize=(8,8))
plt.pie(vehicle_no, labels= label,
       startangle=90,autopct='%.1f %%')
plt.show()
```

```python
from collections import Counter

text= "Introduction and Welcome: The meeting commenced with a warm welcome extended to all attendees. Gratitude was expressed for their keen interest in joining the Environmental Science Club."

def explore_text(text):
    text=text.lower()
    words= text.split()
    num_words=len(words)

    num_words=len(words)
    unique_words=set(words)
    num_unique_words=len(unique_words)
    word_counts=Counter(words)
    most_common_words=word_counts.most_common(5)

    print('Total number of words:', num_words)
    print('Number of unique words: ', num_unique_words)
    print('Most common words: ', most_common_words)

explore_text(text)

```
